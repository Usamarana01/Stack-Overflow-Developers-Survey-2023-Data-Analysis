# Remote work in india

```python
remote_work_india = (indian_dev['RemoteWork'].value_counts() / indian_dev['RemoteWork'].count() * 100).round(2)

# Plotting the bar chart
plt.figure(figsize=(8, 3))
remote_work_india.plot(kind='bar', color='skyblue')
plt.title('Percentage of Remote Work Among Developers in India')
plt.xlabel('Remote Work Status')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# remote work in other countries

```python
(country_grp['RemoteWork'].value_counts().head(10)/(country_grp['RemoteWork'].count()) * 100 ).round(2)
```

```python
Pak_dev=country_grp.get_group('Pakistan')
```

```python
((Pak_dev['RemoteWork'].value_counts())/(Pak_dev['RemoteWork'].count()) * 100 ).round(2)
```

```python
remote_work_pak=((Pak_dev['RemoteWork'].value_counts())/(Pak_dev['RemoteWork'].count()) * 100 ).round(2)
# Plotting the bar chart
plt.figure(figsize=(8, 3))
remote_work_pak.plot(kind='bar', color='green')
plt.title('Percentage of Remote Work Among Developers in Pakistan')
plt.xlabel('Remote Work Status')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
## Comparion between India and Pakistan
```

```python
# Plotting the comparison bar chart
categories = remote_work_india.index
ind = np.arange(len(categories))
width = 0.35

plt.figure(figsize=(8, 6))

plt.bar(ind - width/2, remote_work_india, width, label='India', color='skyblue')
plt.bar(ind + width/2, remote_work_pak, width, label='Pakistan', color='green')

plt.xlabel('Remote Work Status')
plt.ylabel('Percentage')
plt.title('Comparison of Percentage of Remote Work Among Developers in India and Pakistan')
plt.xticks(ind, categories)
plt.legend()
plt.tight_layout()
plt.show()
```

# avg developers age 

```python
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70,
    'Prefer not to say': None  # Assigner la valeur NaN à "Prefer not to say"
}

# Mapper les catégories d'âge aux valeurs numériques
df['NumericAge'] = df['Age'].map(age_mapping)

# Calculer l'âge moyen en excluant les valeurs NaN
mean_age = df['NumericAge'].mean(skipna=True)

age_ranges = {
    (0, 18): "Under 18 years old",
    (18, 25): "18-24 years old",
    (25, 35): "25-34 years old",
    (35, 45): "35-44 years old",
    (45, 55): "45-54 years old",
    (55, 65): "55-64 years old",
    (65, 100): "65 years or older"
}

for age_range, category in age_ranges.items():
    if age_range[0] <= mean_age < age_range[1]:
        print(category)
        break
```
